```python
from pyathena import connect
import pandas as pd
import matplotlib.pyplot as plt
# Establish connection to Athena
conn = connect(s3_staging_dir='s3://cs437s3out/', region_name='us-east-2')

# Query Athena
query = 'SELECT * FROM "kevin-database"."cs436s3" '


df = pd.read_sql(query, conn)

```

```python
import numpy as np

# Assuming df is your DataFrame
# Replace with code to load your DataFrame if needed
# df = pd.read_csv('your_data_file.csv') 

# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['vehicle_speed'], df['vehicle_noise'], color='blue', label='Data points')

# Calculate the best-fit line
m, b = np.polyfit(df['vehicle_speed'], df['vehicle_noise'], 1)  # 1 for linear fit
plt.plot(df['vehicle_speed'], m * df['vehicle_speed'] + b, color='red', label='Best fit line')

# Add titles and labels
plt.title('Vehicle Speed vs. Vehicle Noise with Best Fit Line')
plt.xlabel('Vehicle Speed')
plt.ylabel('Vehicle Noise')
plt.legend()

# Display the plot
plt.show()
```

```python
# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['vehicle_angle'], df['vehicle_co2'], color='blue', label='Data points')

# Calculate the best-fit line
m, b = np.polyfit(df['vehicle_angle'], df['vehicle_co2'], 1)  # 1 for linear fit
plt.plot(df['vehicle_angle'], m * df['vehicle_angle'] + b, color='red', label='Best fit line')

# Add titles and labels
plt.title('Vehicle Angle vs Vehicle CO2 with Best Fit Line')
plt.xlabel('Vehicle Angle')
plt.ylabel('Vehicle CO2')
plt.legend()

# Display the plot
plt.show()
```
